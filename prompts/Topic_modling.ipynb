#Feminism Movement


Topic modelling for feminism dataset.. we extracted the topics for each stance' tweets separately

```python
df = pd.read_csv('Real_train_feminism_664.csv')
df.head()
```

```python
Tweet=df[df['Stance']== 'FAVOR']['Tweet'] #choose the stance' tweets for topic modelling
Tweet.head()
```

```python
#pre-process the tweets before use the LDA model
nltk.download('punkt')
nltk.download('stopwords')
stops = set(stopwords.words("english"))
additional_stopwords = ["doesnt", "im", "dont", "rt","semst","feminist", "feminists"]  # remove specific words in Favor' tweets
#additional_stopwords = ["semst", "feminist", "feminists"]  # remove specific words in Against' tweets
#additional_stopwords = ["semst"]  # remove specific words in None' tweets
stops.update(additional_stopwords)

def preprocess(tweet):
  tokenlist=[]
  tweet = re.sub(r"http\S+|www\S+|https\S+", "", tweet)  # remove URLs
  tweet = re.sub(r"@\w+", "", tweet)  # remove usernames
  tokens = nltk.word_tokenize(tweet) #tokenization
  tokenlist= [w.lower() for w in tokens if w.isalpha()] #make each token in owest case and remove numbers and punctuations

  tokenlist= [word for word in tokenlist if not word in stops] #remove stop words
  return ' '.join([w for w in tokenlist])
```

```python
# apply preprocessing
Tweet= Tweet.apply(preprocess)
Tweet.head()
```

```python
# create a dictionary
dictionary_Tweets = corpora.Dictionary([tweet.split() for tweet in Tweet])
# convert the tweets into a bag-of-words corpus (counts the frequency of each word)
corpus_Tweets=[]
for tweet in Tweet:
  tweet=tweet.split()
  text= dictionary_Tweets.doc2bow(tweet)
  corpus_Tweets.append(text)
print(corpus_Tweets[:1]) #viewing
```

```python
num_topics = 10
# built LDA model
lda_model = gensim.models.LdaModel(corpus_Tweets, num_topics=num_topics, id2word=dictionary_Tweets, passes=10)
```

#Women Empwerment

Topic modelling for Women emperment dataset.. we extracted the topics for each stance' tweets separately

```python
df = pd.read_csv('Real_train_women_1190.csv')
df.head()
```

```python
Tweet=df[df['stance']== 'None']['text'] #choose the stance' tweets for topic modelling
Tweet.head()
```

```python
#pre-process the tweets before use the LDA model

# Download the necessary resources
nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words("arabic"))
additional_stopwords = ["التمكين", "تمكين_المرأة", "يوم_المراه_العالمي","يوم_المرأة_العالمي","للمرأة","","قائدات_التقنية","المرأه","المرأة","MENTION", "تمكين", "ان","الى","اي","URL","الرجل","الله"]
stop_words.update(additional_stopwords)

def preprocess_arabic(tweet):

    tweet = re.sub(r"http\S+|www\S+|https\S+", "", tweet) # emove URLs
    tweet = re.sub(r"@\w+", "", tweet) # remove usernames and mentions
    tweet = re.sub(r"[^\w\s]", "", tweet)  # remove special characters and numbers
    tweet = re.sub(r"\d+", "", tweet)
    tokens = word_tokenize(tweet) # tokenize the tweet
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words] # remove stopwords

    preprocessed_tweet = " ".join(filtered_tokens)

    return preprocessed_tweet
```

```python
# apply preprocessing
Tweet= Tweet.apply(preprocess_arabic)
Tweet.head()
```

```python
# create a dictionary
dictionary_Tweets = corpora.Dictionary([tweet.split() for tweet in Tweet])
# convert the tweets into a bag-of-words corpus (counts the frequency of each word)
corpus_Tweets=[]
for tweet in Tweet:
  tweet=tweet.split()
  text= dictionary_Tweets.doc2bow(tweet)
  corpus_Tweets.append(text)
print(corpus_Tweets[:1]) #viewing
```

```python
num_topics = 5
# perform LDA topic modeling
lda_model = gensim.models.LdaModel(corpus_Tweets, num_topics=num_topics, id2word=dictionary_Tweets, passes=10)
```
