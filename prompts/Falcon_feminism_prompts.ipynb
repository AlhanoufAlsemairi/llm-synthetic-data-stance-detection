```python
# the main code for falon creators can be found in The hugging face hub (https://huggingface.co/tiiuae/falcon-7b-instruct)
#Load the libraries
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model = "tiiuae/falcon-7b-instruct"

#initialize a tokenizer using the model
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
#set generation pipeline using Hugging Face Transformers Pipeline
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
)
```

**We used simple prompts for each stance as explined in the document..**

#Against Stance

```python
prompt = '''generate 20 informal tweets that goes against feminism movements.The tweets may include errors slang, mentions and hashtags. Be offensive and divers.
            separate each tweet by new line.
    '''
    #the prompt was changed slightly for each subset of tweets.
sequences = pipeline(
    prompt,
    max_length=4000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

#FAVOR Stance

```python
prompt = '''USER: Create 20 informal tweets that support feminism. Use variuos tones and be aggressive. your tweets may include mentions and hashtags.
            Each tweet should be in new line.
            ASSISTANT:
    '''
    #the prompt was changed slightly for each subset of tweets.
sequences = pipeline(
    prompt,
    max_length=4000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
